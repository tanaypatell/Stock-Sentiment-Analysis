## Stock Sentiment Analysis using News Headlines

## Introduction

This dataset was obtained from Aaron7sun's Kaggle account. In this notebook, we will attempt to anticipate whether stock prices will grow or decrease by analysing the sentiments of the top 25 headlines of a particular date.

## Importing Necessary Modules

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

## Importing CSV File 

```python
df=pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

#### Let's Dive into our CSV File to get a glance of what we are dealing with

```python
df.head()
```

### Spliting the dataset based on timeline

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

### Data Preprocessing

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)

```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
#predictions = randomclassifier.predict(test_dataset)
```

### Training Models

```python
ran = RandomForestClassifier(n_estimators=200,criterion='entropy')
naive = MultinomialNB()
svm = SVC()
logreg = LogisticRegression()
```

```python
models = {"Random forest" : ran,
          "SVM" : svm,
          "Naive bayes" : naive,
         "Logistic Regression" : logreg}
scores= { }

for key, value in models.items():    
    model = value
    model.fit(traindataset, train['Label'])
    predictions = model.predict(test_dataset)
    scores[key] = model.score(test_dataset, test['Label'])
```

```python
scores_frame = pd.DataFrame(scores, index=["Accuracy Score"]).T
scores_frame.sort_values(by=["Accuracy Score"], axis=0 ,ascending=False, inplace=True)
scores_frame
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x=scores_frame.index,y=scores_frame["Accuracy Score"])
plt.xticks(rotation=90) 
```

## Model Evaluation

### Confusion Matrix and Precision,Recall and F1 Score

```python
from sklearn.metrics import confusion_matrix

final_models = {"Naive bayes" : naive,
                 "Logistic Regression" : logreg,
                   "Random forest" : ran}

for key, value in final_models.items():
    model = value
    y_true = test['Label']
    y_pred = model.predict(test_dataset)
    cm = confusion_matrix(y_true, y_pred)


    f, ax = plt.subplots(figsize=(5,5))
    sns.heatmap(cm,fmt=".0f", annot=True,linewidths=0.2, linecolor="purple", ax=ax)
    plt.title('Confusion Matrix of %s' %key)
    plt.xlabel("Predicted")
    plt.ylabel("Grand Truth")
    plt.show()
    
    
    TN = cm[0,0]
    TP = cm[1,1]
    FN = cm[1,0]
    FP = cm[0,1]
    Precision = TP/(TP+FP)
    Recall = TP/(TP+FN)
    F1_Score = 2*(Recall * Precision) / (Recall + Precision)
    Results = pd.DataFrame([[Precision, Recall, F1_Score]],columns=["Precision", "Recall", "F1 Score"], index=["Results of %s" %key])
    print(Results)
```
